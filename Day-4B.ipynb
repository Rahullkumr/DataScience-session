### Multiple Regression

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('mtcars.csv')
```

```python
df
```

```python
salary = pd.read_csv('Salary_Data.csv')
salary.corr()
```

### Separate the input and output

```python
x = df[['disp', 'hp', 'wt']]
y = df['mpg']
```

```python
x
```

```python
y
```

## Let's do some ML work

### import the class

```python
from sklearn.linear_model import LinearRegression
```

### create the object

```python
reg = LinearRegression()
```

### train the model

```python
reg.fit(x, y)
```

### predict on unseen data

```python
disp = 221
hp = 115
wt = 2.51
reg.predict([[disp, hp, wt]])
```

```python
reg.score(x, y)
```

```python
# 82.6 % accuracy is there in above 
```

### Decission Tree Classification

#### import the dataset

```python
df = pd.read_csv('Social_Network_Ads.csv')
```

```python
df.shape
```

```python
df
```

```python
# 0 means purchased and 1 means not purchased through the ads
# output variable is purchased
# two most imp features are age and salary
# gender is not because it is not having any effect on output variable
```

#### Separate input and ooutput variables

```python
x = df[['Age', 'EstimatedSalary']]
y = df['Purchased']
x;
```

```python
# how many values are present in each category of output variable is known using Series fn ==> value_counts()
```

```python
y.value_counts()
```

#### visualize the data

```python
plt.title('Purchase Data')
plt.xlabel('Age')
plt.ylabel('Salary')
plt.grid()
plt.scatter(x['Age'], x['EstimatedSalary'], c = y); # c = y ==> see whether groups are present 
```

```python
## Let's build the model of decission tree classification
```

#### import the class

```python
from sklearn.tree import DecisionTreeClassifier
```

#### Create the object

```python
classifier = DecisionTreeClassifier(random_state = 0)
```

#### train the model

```python
classifier.fit(x, y)
```

```python
# let's draw the decission tree
```

#### plot the tree

```python
from sklearn.tree import plot_tree
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize = (12, 15))
plot_tree(classifier, fontsize = 7, feature_names = ['age', 'sal'], class_names = ['y', 'n'], filled = True, rounded = True);
```

```python
# not interested in tree, important is whether it can predict or not
```

#### predict on new data

```python
new = pd.DataFrame({
    'Age': [23,45,54,67],
    'EstimatedSalary': [45000, 89000, 23000, 134000]
})
```

```python
new
```

```python
# predict whether above 4 will purchase or not
```

```python
classifier.predict(new) # classifier = object
```
